本函数实现裁剪显示图片中分数的功能

输出4个参数：
1. 分数 6位，
2. 金币 两位，
3. 关卡 1-1，
4. 时间3位，最大400

```python
# 单个图片像素 7*7
# 3个数字像素---23*7
# 总图片像素 256*240

# 数字的竖轴，顶部24位置开始，到31结束
#4个参数：1. 分数 6位，2. 金币 两位，3. 关卡 1-1，4. 时间3位，最大400
```

```python
import cv2
import os

# 设置图片路径
image_path = "picture/output_570.jpg"  # 替换为您的图片文件名

# 读取图片
img = cv2.imread(image_path)

if img is None:
    print(f"无法读取图片：{image_path}")
else:
    # 设置裁剪坐标 (x1, y1) 到 (x2, y2)
    x1, y1 = 12, 23
    x2, y2 = 232, 32
    
    # 裁剪图片 (注意：OpenCV使用[y:y+h, x:x+w]格式)
    cropped_img = img[y1:y2, x1:x2]
    # 保存裁剪后的图片
    cv2.imwrite("cropped_result.jpg", cropped_img)
```
